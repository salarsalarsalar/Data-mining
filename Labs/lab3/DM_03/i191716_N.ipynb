## Low variance filter

```python
data1 = data.drop('class', axis=1)
```

```python
data1.isnull().sum()
```

```python
data1.replace(np.NaN,0,inplace = True)
```

```python
normalize_data = normalize(data1)
```

```python
scale_data = pd.DataFrame(normalize_data)
scale_data.var()
```

```python
#storing the variance and name of variables

variance = scale_data.var()
columns = data1.columns
```

```python
#saving the names of variables having variance more than a threshold value

df = [ ]

for i in range(0,len(variance)):
    if variance[i] >= 0.00035: #setting the threshold as 1%
        df.append(columns[i])
df
```

```python
# creating a new dataframe using the above variables
data2 = data1[df]
data2
```

## High Correlation Filter

```python
data2.astype(float)
correlation = data2.corr()
numeric_columns = data2.columns

high_corr = [ ]

for c1 in numeric_columns:
  for c2 in numeric_columns:
    if c1 != c2 and c2 not in high_corr and correlation[c1][c2] > 0.9:
      high_corr.append(c1)

data2=data2[high_corr]
data2
```
