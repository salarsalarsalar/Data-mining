```python
label_encoder = LabelEncoder()
```

```python
df['Gender']= label_encoder.fit_transform(df['Gender'])
df['STATE']= label_encoder.fit_transform(df['STATE'])
df['Previous_Degree']= label_encoder.fit_transform(df['Previous_Degree'])

df['Marital_status']= label_encoder.fit_transform(df['Marital_status'])
df['Place_you_belong_to']= label_encoder.fit_transform(df['Place_you_belong_to'])
df['perceived#Job#Skill']= label_encoder.fit_transform(df['perceived#Job#Skill'])

df['Specialization']= label_encoder.fit_transform(df['Specialization'])
```

```python
df
```

```python
X = df.iloc[:, :-1].values
y = df['Specialization'].values
```

```python
labels, unique = pd.factorize(y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size = 0.40)
```

```python
def KNNClassify(X_test, Y_train = y_train,X_train = X_train, k = 8):
   
    min_dist = []
    for i,point in enumerate(X_train):
        d = []
        for i in range(0,3):
            d.append((point[0]-X_test[0])**2)        
        d_a = sum(d)
        dist = np.sqrt(d_a)
        min_dist.append((i,dist))
    min_dist.sort(key = takeSecond)
    
    neighbours = min_dist[:k]
    idx = []
    for tup in neighbours:
        idx.append(tup[0])
    output = Y_train[idx]
    values, counts = np.unique(output, return_counts=True)
    max_idx = np.argmax(counts)
    return values[max_idx]
```

```python
def takeSecond(elem):
    return elem[1]
```

```python
predictions = list(map(KNNClassify, X_test))
```

```python
def accuracy(pred , y_test):
    count = 0
    for i in range(len(pred)):
        if pred[i] == y_test[i]:
            count +=1
    acc = count/len(pred)
    return acc
```

```python
accuracy(predictions, y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(X_train,y_train)
y_pred = knn.predict(X_test)
knn.score(X_test,y_test)
```
